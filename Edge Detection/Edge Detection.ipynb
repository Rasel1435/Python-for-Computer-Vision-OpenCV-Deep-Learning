```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import cv2

%matplotlib inline
```

```python
def display(img, cmap="gray"):
    fig = plt.figure(figsize=(12,10))
    ax = fig.add_subplot(111)
    ax.imshow(img, cmap="gray")
```

```python
img = cv2.imread(".\\images\dog.jpg")
```

```python
flip_img = cv2.flip(img,1)
cv2.imwrite(".\\images\dog.jpg", flip_img)
```

```python
display(flip_img)
```

```python
edges = cv2.Canny(image=flip_img, threshold1=127,threshold2=127)
display(edges)
```

```python
edges = cv2.Canny(image=flip_img, threshold1=0,threshold2=255)
display(edges)
```

```python
med_val = np.median(flip_img)
med_val
```

```python
lower = int(max(0,0.7*med_val))
upper = int(min(255,1.3*med_val))
```

```python
edges = cv2.Canny(image=flip_img, threshold1=lower,threshold2=upper+100)
display(edges)
```

```python
blured_img = cv2.blur(flip_img, ksize=(5,5))
```

```python
edges = cv2.Canny(image=blured_img, threshold1=lower,threshold2=upper+50)
display(edges)
```
