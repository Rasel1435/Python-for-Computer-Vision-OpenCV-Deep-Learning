```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import cv2

%matplotlib inline
```

```python
def display(img, cmap="gray"):
    fig = plt.figure(figsize=(12,10))
    ax = fig.add_subplot(111)
    ax.imshow(img, cmap="gray")
```

```python
nadia = cv2.imread(".\\images\\nadia_murad.jpg", 0)
denis = cv2.imread(".\\images\\Denis_Mukwege.jpg", 0)
solvay = cv2.imread(".\\images\\Solvay_conference_1927.jpg", 0)
```

```python
display(solvay)
```

```python
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + "haarcascade_frontalface_default.xml")
```

```python
def detect_face(img):
    face_img = img.copy()
    face_rects = face_cascade.detectMultiScale(face_img)

    for (x,y,w,h) in face_rects:
        cv2.rectangle(face_img,(x,y),(x+w,y+h),(255,255,255), 10)
        
    return face_img
```

```python
result = detect_face(denis)
result2 = detect_face(nadia)
result3 = detect_face(solvay)
```

```python
display(result)
```

```python
def detect_face(img):
    face_img = img.copy()
    face_rects = face_cascade.detectMultiScale(face_img, scaleFactor=1.2, minNeighbors=1)

    for (x,y,w,h) in face_rects:
        cv2.rectangle(face_img,(x,y),(x+w,y+h),(255,255,255), 10)
        
    return face_img
```

```python
result3 = detect_face(solvay)
```

```python
display(result3)
```

```python
eye_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + "haarcascade_eye.xml")
```

```python
def detect_eyes(img):
    face_img = img.copy()
    eyes_rects = eye_cascade.detectMultiScale(face_img, scaleFactor=1.2, minNeighbors=20)

    for (x,y,w,h) in eyes_rects:
        cv2.rectangle(face_img,(x,y),(x+w,y+h),(255,255,255), 10)
        
    return face_img
```

```python
result = detect_eyes(nadia)
display(result)
```

```python
result = detect_eyes(denis)
display(result)
```

```python
cap = cv2.VideoCapture(0)

while True:
    ret, frame = cap.read(0)
    frame = detect_face(frame)
    cv2.imshow("Video Face Detect", frame)

    k = cv2.waitKey(1)
    if k == 27:
        break
cap.release()
cv2.destroyAllWindows()
```
