```python
gorila = cv2.imread(".\images\gorila.jpg",0)
```

```python
def display_gorila(img,cmap=None):
    fig = plt.figure(figsize=(10,8))
    ax = fig.add_subplot(111)
    ax.imshow(img,cmap)

display_gorila(gorila,cmap="gray")
```

```python
gorila.shape
```

```python
hist_values = cv2.calcHist([gorila],channels=[0],mask=None,histSize=[256],ranges=[0,256])
```

```python
plt.plot(hist_values)
```

```python
eq_gorilla = cv2.equalizeHist(gorila)
display_gorila(eq_gorilla,cmap="gray")
```

```python
hist_values = cv2.calcHist([eq_gorilla],channels=[0],mask=None,histSize=[256],ranges=[0,256])
```

```python
plt.plot(hist_values)
```

```python
color_gorila = cv2.imread(".\images\gorila.jpg")
show_gorila = cv2.cvtColor(color_gorila,cv2.COLOR_BGR2RGB)
```

```python
display_gorila(show_gorila)
```
