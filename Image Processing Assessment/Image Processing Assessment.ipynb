```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import cv2

%matplotlib inline
```

```python
def display_img(img, cmap=None):
    fig = plt.figure(figsize=(10,8))
    ax = fig.add_subplot(111)
    ax.imshow(img,cmap)
```

```python
giraffe = cv2.imread(".\images\giraffs.jpg")
giraffe = cv2.cvtColor(giraffe,cv2.COLOR_BGR2RGB)
```

```python
display_img(giraffe)
```

Apply a binary threshold onto the image

```python
giraffe = cv2.imread(".\images\giraffs.jpg",0)
ret, thres1 = cv2.threshold(giraffe,127,255,cv2.THRESH_BINARY)
display_img(thres1, cmap="gray")
```

HSV COLOR

```python
giraffe = cv2.imread(".\images\giraffs.jpg")
giraffe = cv2.cvtColor(giraffe,cv2.COLOR_BGR2HSV)
display_img(giraffe)
```

```python
kernel = np.ones(shape=[4,4],dtype=np.float32)/10
kernel
```

```python
giraffe = cv2.imread(".\images\giraffs.jpg")
giraffe = cv2.cvtColor(giraffe,cv2.COLOR_BGR2RGB)
```

```python
result = cv2.filter2D(giraffe,-1,kernel)
display_img(result)
```

sobel Filter

```python
giraffe = cv2.imread(".\images\giraffs.jpg",0)
sobelx = cv2.Sobel(giraffe,cv2.CV_64F,1,0,ksize=5)
display_img(sobelx, cmap="gray")
```

Hidtogram Red, Blue, Green

```python
giraffe = cv2.imread(".\images\giraffs.jpg")
color = ['b','g','r']
plt.figure(figsize=(18,15))
for i, col in enumerate(color):
    hist_values = cv2.calcHist([giraffe],[i],None,[256],[0,256])
    plt.plot(hist_values,color=col)
plt.title("MY HISTOGRAM")
```
