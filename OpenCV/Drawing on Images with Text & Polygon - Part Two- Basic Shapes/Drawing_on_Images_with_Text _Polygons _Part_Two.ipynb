```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from PIL import ImageFont, ImageDraw, Image
import cv2
import time
from tkinter import font
%matplotlib inline
```

```python
blank_img = np.zeros(shape=(512,512,3), dtype=np.int16)
```

```python
blank_img.shape
```

```python
type(blank_img)
```

```python
plt.imshow(blank_img)
```

```python
cv2.line(blank_img, pt1=(0,0), pt2=(512,512), color=(0,255,0), thickness=5)
plt.imshow(blank_img)
```

```python
cv2.rectangle(blank_img, pt1=(400,156), pt2=(350,200), color=(255,0,0), thickness=5)
plt.imshow(blank_img)
```

```python
cv2.circle(blank_img, center=(150,356), radius=50, color=(255,0,0), thickness=-1)
plt.imshow(blank_img)
```

```python
cv2.putText(blank_img, text='Rasel', org=(10, 500), fontFace=4,
            fontScale=4, color=(255, 255, 255), thickness=3, lineType=cv2.LINE_AA)
plt.imshow(blank_img)
```
